```python
def get_sent(sents,i):
    return tuple([sents[x] if x <len(sents) else " " for x in range(0,i)])

def get_sents(sents):
    return sents[0],get_sent(sents,1),get_sent(sents,2),get_sent(sents,3),get_sent(sents,4),get_sent(sents,5), get_sent(sents,6), get_sent(sents,7),get_sent(sents,8),get_sent(sents,9)," ".join(sents[10:])
        
```

```python
import re
def clean_text(text):
    text = re.sub('[0-9]','',text)
    text = text.replace(","," , ")
    text = text.replace(":"," : ")
    text = text.replace("!"," ! ")
    text = text.replace("?"," ? ")
    text = text.replace(";", " ; ")
    return text
```

```python
def prepare_data(sents,tokenizer=tokenizer,maxlen=30):
    sents = get_sent(sents,5)
    inputs = []
    input = []
    for sent in sents:
        val = tf.keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([sent]),maxlen)
        input.append(val[0])
    
    inputs.append(input)
    return np.array(inputs)
    
    
```

```python
def get_attn_sentences_sorted(sents,attmodel,maxlen=50):
    sentences_result = np.argsort(attmodel.predict(prepare_data(sents)),axis=1)[::-1]
    return sentences_result[0]
    #out = list(zip(sentences_result,sents))
    #ax = np.argsort(sentences_result)[::-1]
    #print(ax)
    #return [sents[i] for i in ax]
```

```python
sent_index = {0:'activation_36',
             1:'activation_37',
             2:'activation_38',
             3:'activation_39',
             4:'activation_40',
             }
```

```python
rev_word_index = {j:i for i,j in tokenizer.word_index.items()}
rev_word_index[0] = "<blank>"
def get_attn_words(sents,layer,attmodel,rev_word_index,index,maxlen=50):
    word_attn_model = tf.keras.models.Model(inputs=attmodel.input,outputs=attmodel.get_layer(layer).output)
    input = prepare_data(sents)
    input = input[0][index]
    words_result = np.argsort(word_attn_model.predict(prepare_data(sents))[0])[::-1]
    #print(words_result)
    #print(rev_word_index[input[np.argmax(word_attn_model.predict(prepare_data(sents)),axis=1)][0]])
    return [input[x] for x in words_result]
```
