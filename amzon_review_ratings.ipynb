```python
import re
def clean_text(text):
    text = re.sub('[0-9]','',text)
    text = text.replace(","," , ")
    text = text.replace(":"," : ")
    text = text.replace("!"," ! ")
    text = text.replace("?"," ? ")
    text = text.replace(";", " ; ")
    text = text.lower()
    return text
```

```python
def load_glove_embedding(glove_path):
    word2emb = {}
    with open(glove_path, 'rb') as fglove:
        for line in fglove:
            cols = line.strip().split()
            word = cols[0]
            embedding = np.array(cols[1:], dtype='float32')
            word2emb[word] = embedding
    return word2emb

def prepare_embedding_matrix(embedding, tokenizer, emb_dim=100):
    matrix = np.zeros((len(tokenizer.word_index)+1,emb_dim))
    count = 0
    absent_words = []
    for key in tokenizer.word_index:
        if str.encode(key.replace("'", "").replace('"','')) in embedding.keys() :
            matrix[tokenizer.word_index[key]] = embedding[str.encode(key.replace("'", "").replace('"',''))]
        else:
            count+=1
            absent_words.append(key)
    return matrix
```

```python
def prepare_data(sents,tokenizer=tokenizer,maxlen=30):
    sents = get_sent(sents,3)
    inputs = []
    input = []
    for sent in sents:
        val = tf.keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([sent]),maxlen)
        input.append(val[0])
    
    inputs.append(input)
    return np.array(inputs)
    
    
```

```python
def get_attn_sentences_sorted(sents,attmodel,maxlen=50):
    result = attmodel.predict(prepare_data(sents))[0]
    sentences_result = np.argsort(result)[::-1]
    return tuple(zip(sentences_result, np.sort(result)[::-1]))
    #out = list(zip(sentences_result,sents))
    #ax = np.argsort(sentences_result)[::-1]
    #print(ax)
    #return [sents[i] for i in ax]
```

```python
rev_word_index = {j:i for i,j in tokenizer.word_index.items()}
rev_word_index[0] = "<blank>"
def get_attn_words(sents,layer,attmodel,rev_word_index,index,maxlen=50):
    word_attn_model = tf.keras.models.Model(inputs=attmodel.input,outputs=attmodel.get_layer(layer).output)
    input = prepare_data(sents)
    input = input[0][index]
    probs = word_attn_model.predict(prepare_data(sents))[0]
    words_result = np.argsort(word_attn_model.predict(prepare_data(sents))[0])[::-1]
    #print(words_result)
#     print([input[x] for x in words_result])
    return [(input[x],probs[x]) for x in words_result]
```
