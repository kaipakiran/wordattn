```python
import pandas as pd
```

```python
from nltk import sent_tokenize
```

```python
def split_text(text):
    l = sent_tokenize(str(text))
    return l
```

```python
df = pd.read_csv("./data/imdb_train.csv")
```

```python
df['comment_list'] = df['comments'].apply(split_text)
```

```python
df['comment_list_len'] = df['comment_list'].apply(len)
```

```python
import numpy as np
np.mean(df['comment_list_len'])
```

```python
def get_sent(sents,i):
    return sents[i] if len(sents)>i else " "
def get_sents(sents):
    return sents[0],get_sent(sents,1),get_sent(sents,2),get_sent(sents,3),get_sent(sents,4),get_sent(sents,5), get_sent(sents,6), get_sent(sents,7),get_sent(sents,8),get_sent(sents,9)," ".join(sents[10:])
        
```

```python
new_df = pd.DataFrame(df['comment_list'].apply(get_sents).tolist(),columns=['sent1','sent2','sent3','sent4','sent5','sent6','sent7','sent8','sent9','sent10','sent11'])
```

```python
import re
def clean_text(text):
    text = re.sub('[0-9]','',text)
    text = text.replace(","," , ")
    text = text.replace(":"," : ")
    text = text.replace("!"," ! ")
    text = text.replace("?"," ? ")
    text = text.replace(";", " ; ")
    return text
```

```python
for col in new_df.columns:
    new_df[col] = new_df[col].apply(clean_text)
```

```python
tokenizer= tf.keras.preprocessing.text.Tokenizer()
```

```python
tokenizer.fit_on_texts(df['comments'])
```

```python
def get_seq(col,len=50):
    q = list(col)
    return tf.keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences(q),50)
```

```python
seq1 = get_seq(new_df['sent1'])
seq2 = get_seq(new_df['sent2'])
seq3 = get_seq(new_df['sent3'])
seq4 = get_seq(new_df['sent4'])
seq5 = get_seq(new_df['sent5'])
seq6 = get_seq(new_df['sent6'])
seq7 = get_seq(new_df['sent7'])
seq8 = get_seq(new_df['sent8'])
seq9 = get_seq(new_df['sent9'])
seq10 = get_seq(new_df['sent10'])
seq11 = get_seq(new_df['sent11'])
```

```python
def prepare_data(sents,tokenizer=tokenizer,maxlen=50):
    sents = get_sents(sents)
    inputs = []
    input = []
    for sent in sents:
        val = tf.keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([sent]),maxlen)
        input.append(val[0])
    
    inputs.append(input)
    return np.array(inputs)
    
    
```

```python
def get_attn_sentences_sorted(sents,attmodel,maxlen=50):
    sentences_result = np.argsort(attmodel.predict(prepare_data(sents)),axis=1)[::-1]
    return sentences_result[0]
    #out = list(zip(sentences_result,sents))
    #ax = np.argsort(sentences_result)[::-1]
    #print(ax)
    #return [sents[i] for i in ax]
```

```python
sent_index = {0:'activation_12',
             1:'activation_13',
             2:'activation_14',
             3:'activation_15',
             4:'activation_16',
             5:'activation_17',
             6:'activation_18',
             7:'activation_19',
             8:'activation_20',
             9:'activation_21',
             10:'activation_21'}
```

```python
rev_word_index = {j:i for i,j in tokenizer.word_index.items()}
rev_word_index[0] = "<blank>"
def get_attn_words(sents,layer,attmodel,rev_word_index,index,maxlen=50):
    word_attn_model = tf.keras.models.Model(inputs=attmodel.input,outputs=attmodel.get_layer(layer).output)
    input = prepare_data(sents)
    input = input[0][index]
    words_result = np.argsort(word_attn_model.predict(prepare_data(sents)),axis=1)[::-1][0]
    return [input[x] for x in words_result]
```
